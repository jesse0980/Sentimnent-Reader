```python
import os
import json
import time as t
import numpy as np
import glob
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from textblob import TextBlob
import csv
import plotly.express as px
import pandas_datareader as web
import plotly.graph_objects as go
```

```python
stock = 'ETH-USD'
 
df = web.DataReader(stock, data_source='yahoo', start='01-01-2019')

```

```python
trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'MSFT',
    'showlegend': True
}
```

```python
data = [trace1]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'Microsoft(MSFT) Moving Averages',
        'font': {
            'size': 15
        }
    }
})
```

```python
fig = go.Figure(
    data,
    layout_title_text="ETH"
)
fig.show()
```

```python
def plotScores(term):
        df = pd.read_csv("{}.csv".format(term))
        print(df.head())
        # pd.to_datetime(df['date'])
        df.plot(kind = 'scatter',
                x = 'DateTime',
                y = 'Score',
                color = 'red')
        plt.show(block=False)
        fig = px.line(df, x = 'DateTime', y = 'Score', title=term)
        fig.show()
        fig.write_html("rando.html")
plotScores("bitcoin")
```
